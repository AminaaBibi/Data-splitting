```python

import pandas as pd
import numpy as np
df = pd.read_csv('ECG200_TRAIN.csv', delimiter='  ', header = None)
df
```

```python
df2 = df[df[0] == -1.0]
df3 = df.append(df2)
df3 = df3.sample(frac=1).reset_index(drop=True) # shuffle the dataframe
df3



```

```python
target = df3.iloc[:,0]
target.value_counts()


```

```python
df4 = df3[df3[0]==1.0]
df5 = df3[df3[0] == -1.0]
```

```python
# take first 15 rows from df5
df6_1 = df5.head(15)
df7_1 = df5.iloc[16:31]
df8_1 = df5.iloc[32:48]
df9_1 = df5.iloc[49:64]
# take 17 rows from df4
df61 = df4.head(17)
df71 = df4.iloc[18:35]
df81 = df4.iloc[36:53]
df91 = df4.iloc[54:72]
# concatenate all the dataframes
chunk1 = pd.concat([df6_1, df61])
chunk2 = pd.concat([df7_1, df71])
chunk3 = pd.concat([df8_1, df81])
chunk4 = pd.concat([df9_1, df91])
# shuffle all the chunks
chunk1 = chunk1.sample(frac=1).reset_index(drop=True)
chunk2 = chunk2.sample(frac=1).reset_index(drop=True)
chunk3 = chunk3.sample(frac=1).reset_index(drop=True)
chunk4 = chunk4.sample(frac=1).reset_index(drop=True)
```

```python
train = pd.concat([chunk1, chunk3, chunk4])
# make X_train and y_train
X_train = train.iloc[:,1:]
y_train = train.iloc[:,0]
# make X_test and y_test
X_test = chunk2.iloc[:,1:]
y_test = chunk2.iloc[:,0]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
pred=clf.predict(X_test)
clf.score(X_test,y_test)
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
```

```python
from sklearn.model_selection import KFold 
X = df3.iloc[:,1:]
y = df3.iloc[:,0]
kf = KFold(n_splits=4, random_state=None)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(X_train, y_train)
    pred=clf.predict(X_test)
    clf.score(X_test,y_test)
    print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
    print("Report : \n", classification_report(y_test, pred))
    print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
```
