```python
import numpy as np
import math
import itertools as it
red = 4
white = 6
blue = 5
total = red + white + blue
sample = list(it.product(['red', 'white', 'blue'], repeat=3))
print(sample)
def prob(n, r):
    return math.factorial(n) / (math.factorial(r) * math.factorial(n - r))

# count the number of times the color appears in the sample
def count(sample):
    red_count = []
    white_count = []
    blue_count = []
    for i in sample:
        red_count.append(i.count('red'))
        white_count.append(i.count('white'))
        blue_count.append(i.count('blue'))
    return red_count, white_count, blue_count
count(sample)


total2 = prob(15, 3)
print(total2)





```

```python
def find_prob(sample):
    red_count, white_count, blue_count = count(sample)
    red_prob = []
    white_prob = []
    blue_prob = []
    for i in red_count:
        red_prob.append(prob(red, i))
    for i in white_count:
        white_prob.append(prob(white, i))
    for i in blue_count:
        blue_prob.append(prob(blue, i))
    return red_prob, white_prob, blue_prob
# find_prob(sample)
```

```python
def find_prob2(sample):
    red_prob, white_prob, blue_prob = find_prob(sample)
    red_prob2 = []
    white_prob2 = []
    blue_prob2 = []
    for i in range(len(red_prob)):
        red_prob2.append(red_prob[i] * white_prob[i] * blue_prob[i] / total2)
    for i in range(len(white_prob)):
        white_prob2.append(red_prob[i] * white_prob[i] * blue_prob[i] / total2)
    for i in range(len(blue_prob)):
        blue_prob2.append(red_prob[i] * white_prob[i] * blue_prob[i] / total2)
    return red_prob2, white_prob2, blue_prob2
find_prob2(sample)

```

```python
def find_max(sample):
    red_prob2, white_prob2, blue_prob2 = find_prob2(sample)
    max_prob = max(red_prob2)
    max_index = red_prob2.index(max_prob)
    return sample[max_index] , max_prob
find_max(sample)

```

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df
```

```python
# seperate the data into 3 different dataframes with all the data for each variety
setosa = df[df['variety'] == 'Setosa']
versicolor = df[df['variety'] == 'Versicolor']
virginica = df[df['variety'] == 'Virginica']
setosa
```

```python
def find_mean_cov(df):
    mean = df.mean()
    cov = df.cov()
    return mean, cov
setosa_mean, setosa_cov = find_mean_cov(setosa)
versicolor_mean, versicolor_cov = find_mean_cov(versicolor)
virginica_mean, virginica_cov = find_mean_cov(virginica)
print(setosa_mean, setosa_cov)
```

```python
from scipy.stats import multivariate_normal
multivariate_normal.pdf([5.1, 3.5, 1.4, 0.2], mean=setosa_mean, cov=setosa_cov)
```

```python
setosa_prob = len(setosa) / len(df)
versicolor_prob = len(versicolor) / len(df)
virginica_prob = len(virginica) / len(df)
print(setosa_prob, versicolor_prob, virginica_prob)
```

```python
# find mutilvariate normal distribution for each variety
setosa_normal = multivariate_normal.pdf([5.1, 3.5, 1.4, 0.2], mean=setosa_mean, cov=setosa_cov)
versicolor_normal = multivariate_normal.pdf([5.1, 3.5, 1.4, 0.2], mean=versicolor_mean, cov=versicolor_cov)
virginica_normal = multivariate_normal.pdf([5.1, 3.5, 1.4, 0.2], mean=virginica_mean, cov=virginica_cov)
print(setosa_normal, versicolor_normal, virginica_normal)
```

```python
setosa_normal = setosa_normal * setosa_prob
versicolor_normal = versicolor_normal * versicolor_prob
virginica_normal = virginica_normal * virginica_prob
print(setosa_normal, versicolor_normal, virginica_normal)
```
