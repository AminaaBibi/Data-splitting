```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Take a random vector in real values
x = np.random.randint(3,5,2)
print("x= ",x)
y = (np.array([x[1],-x[0]]))
print("y= ",y)
print(int(x.dot(y)))
```

```python
# plot the orthogonal vector
plt.plot([0,x[0]],[0,x[1]],'r')
plt.plot([0,y[0]],[0,y[1]],'b')
plt.show()
```

```python
iris = np.genfromtxt('iris.csv', delimiter=',')
iris = iris[~np.isnan(iris).any(axis=1)]
# Iris data to zero-centered data
iris = iris - np.mean(iris, axis=0)
cov = np.cov(iris.T)
eig_val, eig_vec = np.linalg.eig(cov)
print("eigenvalues: ",eig_val)
print("eigenvectors: ",eig_vec)
# scatter plot the data after zero centering
plt.scatter(iris[:,0],iris[:,1])
plt.show()
```
