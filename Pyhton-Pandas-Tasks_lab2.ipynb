```python
# Given a list of strings, return the count of the number of strings where the string length is 2 or more and the first and last chars of the string are the same.

def string_match(words):
    count = 0
    for i in words:
        if len(i)>=2 and i[0]==i[-1]:
            count+=1
    return count

print(string_match(["mango","apple","aba"]))





```

```python
# Given a list of strings, return a list with the strings in sorted order, except group all the strings that begin with &#39;x&#39; first.

def order(words):
    x = []
    y = []
    for i in words:
        if i[0]=="x":
            x.append(i)
        else:
            y.append(i)
    return sorted(x)+sorted(y)

print(order(["xylo", "xylophone", "xyz", "xy", "mango", "apple", "orange", "google"]))
```

```python
# Given a list of non-empty tuples, return a list sorted in increasing order by the last element in each tuple.

def sort_last(list):
    for i in range(len(list)):
        for j in range(len(list)-1):
            if list[j][-1]>list[j+1][-1]:
                list[j],list[j+1] = list[j+1],list[j]
    return list


```

```python
print(sort_last([(8,9),(2,11),(4,5),(10,1),(6,7)]))
```

```python
# Given a list of numbers, return a list where all adjacent == elements have been reduced to a single element, so [1, 2, 2, 3] returns [1, 2, 3].

def remove(nums):
    for i in range(len(nums)-2):
        if nums[i]==nums[i+1]:
            nums.pop(i+1)
    return nums  
```

```python
print(remove([1, 2, 2, 3]))
```

```python
# two lists sorted in increasing order, create and return a merged list of all the elements in sorted order

def merge(list1,list2):
    return sorted(list1+list2)

print(merge([4,5,6],[1,2,3]))
```
